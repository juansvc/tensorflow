```python
import pandas as pd
cols = ['log', 'clasificacion']
df = pd.read_csv('clasification.csv', index_col=False, sep=',', names=cols);
df.sample(n=10)
```

```python
df.dropna(inplace = True)
```

```python
new = df['log'].str.split(' ', n = 1, expand = True)
```

```python
df['username'] = new[0]
```

```python
df['nombre'] = new[1]
```

```python
df.drop(columns =['log'], inplace = True)
```

```python
df.drop(columns =['username'], inplace = True)
```

```python
new2 = df['nombre'].str.split(',', n = 1, expand = True)
```

```python
df['nombre_usuario'] = new2[0]
```

```python
df['ya'] = new2[1]
```

```python
df
```

```python
df.drop(columns =['nombre'], inplace = True)
```

```python
new3 = df['ya'].str.split(',', n = 17, expand = True)
df['a'] = new3[0]
df['b'] = new3[1]
df['c'] = new3[2]
df['d'] = new3[3]
df['e'] = new3[4]
df['f'] = new3[5]
df['g'] = new3[6]
df['h'] = new3[7]
df['i'] = new3[8]
df['j'] = new3[9]
df['k'] = new3[10]
df['l'] = new3[11]
df['m'] = new3[12]
df['n'] = new3[13]
df['o'] = new3[14]
df['p'] = new3[15]
df['q'] = new3[16]
df['r'] = new3[17]
df
```

```python
df.drop(columns =['a','b', 'c','d','e','f','g','h','i','j','k','l','m'], inplace = True)
```

```python
df
```

```python
con_curso = df
con_curso
```

```python
a = df[df['n'].str.contains("course id")]
con_curso['curso'] = a['n']
con_curso.drop(columns =['n'], inplace = True)
```

```python
a = df[df['o'].str.contains("course id")]
con_curso['curso1'] = (a['o'])
con_curso.drop(columns =['o'], inplace = True)
```

```python
a = df[df['p'].str.contains("course id", na=False)]
con_curso['curso2'] = (a['p'])
con_curso.drop(columns =['p'], inplace = True)
```

```python
a = df[df['q'].str.contains("course id", na=False)]
con_curso['curso3'] = (a['q'])
con_curso.drop(columns =['q'], inplace = True)
```

```python
a = df[df['r'].str.contains("course id", na=False)]
con_curso['curso4'] = (a['r'])
con_curso.drop(columns =['r'], inplace = True)
```

```python
con_curso.drop(columns =['ya'], inplace = True)
```

```python
con_curso['cursos'] = con_curso['curso'].map(str) + con_curso['curso1'].map(str) + con_curso['curso2'].map(str) + con_curso['curso3'].map(str) + con_curso['curso4'].map(str)
```

```python
con_curso.drop(columns =['curso','curso1','curso2','curso3','curso4'], inplace = True)
con_curso
```

```python
con_curso = con_curso[con_curso['cursos'].str.contains("course id")]
```

```python
con_curso['cursos1'] = con_curso['cursos'].map(lambda x: x.lstrip('nan'))
```

```python
con_curso['cursos1'] = con_curso['cursos1'].map(lambda x: x.lstrip('course id'))
```

```python
con_curso.drop(columns =['cursos'], inplace = True)
```

```python
con_curso['cursos1'] = con_curso['cursos1'].map(lambda x: x.rstrip('nan'))
con_curso
```

```python
new4 = con_curso['cursos1'].str.split(',', n = 1, expand = True)
con_curso['a'] = new4[0]
con_curso['b'] = new4[1]
```

```python
con_curso.drop(columns =['b','cursos1'], inplace = True)
con_curso = con_curso[con_curso['a'].str.contains("course id")]
```

```python
new5 = con_curso['a'].str.split(' ', n = 6, expand = True)
con_curso['b'] = new5[0]
con_curso['c'] = new5[1]
con_curso['d'] = new5[2]
con_curso['e'] = new5[3]
con_curso['f'] = new5[4]
con_curso['g'] = new5[5]
con_curso['h'] = new5[6]
con_curso.drop(columns =['h'], inplace = True)
```

```python
con_curso['cursos'] = con_curso['b'].map(str) + " " + con_curso['c'].map(str) + " " + con_curso['d'].map(str) + " " + con_curso['e'].map(str) + " " + con_curso['f'].map(str) + " " + con_curso['g'].map(str)
```

```python
con_curso.drop(columns =['a', 'b', 'c', 'd', 'e', 'f' , 'g'], inplace = True)
con_curso
```
