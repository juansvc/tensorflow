```python
import pandas as pd
df1 = pd.read_csv('aprobados.csv', index_col=False, sep=',');
df2 = df1
df1.drop(columns =['id','download_url','key','grade','distinction','verify_uuid','download_uuid','name','created_date','modified_date','error_reason','mode'], inplace = True)
efhe = df1[df1['course_id'].str.contains("EFHE")]
df1['efhe'] = efhe['course_id']
df1 = df1.dropna()
df1.drop(columns =['course_id'], inplace = True)
df2.sample(n=10)
```

```python
df1['efhe2'] = df1['efhe'].str.replace('\W',' ')
df1.sample(n=10)
```

```python
df1['efhe3'] = df1['efhe2'].str.replace('_',' ')
df1.sample(n=10)
```

```python
df1['efhe3'] = df1['efhe3'].str.lower()
df1.drop(columns =['efhe','efhe2'], inplace = True)
df1.sample(n=10)
```

```python
df1.to_csv('aprobadosfianl.csv')
```
